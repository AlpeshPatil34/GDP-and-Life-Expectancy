```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
df.head()
```

```python
df.rename(columns={'Life expectancy at birth (years)' : 'LEABY'},inplace=True)
df = df.replace(['United States of America'], 'USA')
```

```python
plt.bar(df['Country'], df['GDP'], color='green')
plt.title('GDP Comparison of Countries')
plt.xlabel('Countries')
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.savefig('Country_GDP_Comapre.png')
plt.show()
```

```python
plt.bar(df['Country'], df['LEABY'], color = 'green')
plt.title('Comparative bar chart of countries Life Expectancy')
plt.xlabel('Countries')
plt.ylabel('Life Expectancy in years')

plt.savefig('Comparison_of_Life_Expectancy.png')

plt.show()
```

```python
chil_zim = df[(df.Country == 'Chile') | (df.Country == 'Zimbabwe')].reset_index()

plt.bar(chil_zim['Country'], chil_zim['GDP'], color = 'green')
plt.title("Comparative bar chart of Chile's and Zimbabwe's GDP")
plt.xlabel('Countries')
plt.ylabel('GDP in Trillions of U.S. Dollars')

plt.savefig('Chille_Zimbabwe_GDP_Compare.png')

plt.show()
```

```python
f, ax = plt.subplots(figsize=(20, 10))

ax = sns.barplot(x='Country', y='GDP', data=chil_zim, hue='Year', palette='rocket')

ax.set_title('GDP for Chile and Zimbabwe over the years')
ax.set_xlabel('Countries')
ax.set_ylabel('GDP in Trillions of U.S. Dollars')

plt.savefig('GDP_Years_Chile_Zimbabwe.png')

plt.show()
```

```python
fig = plt.subplots(figsize=(15, 10))

ax = sns.violinplot(df['Country'], df['LEABY'])
ax.set_title('Report on Life Expectancy per Country')
ax.set_xlabel('Countries')
ax.set_ylabel('Life Expectancy in years')

plt.savefig('Violin_Plot_Life_Expectancy.png')

plt.show()
```

```python
f, ax = plt.subplots(figsize=(20, 10))

ax = sns.barplot(x='Country', y='GDP', data=df, hue='Year', palette='rocket_r')

ax.set_title('GDP per Country over the years')
ax.set_xlabel('Countries')
ax.set_ylabel('GDP in Trillions of U.S. Dollars')

plt.savefig('GDP_Over_Years.png')

plt.show()
```

```python
f, ax = plt.subplots(figsize=(20, 10)) 

ax = sns.barplot(x='Country', y='LEABY', data=df, hue='Year', palette='rocket')

ax.set_title('Life Expectancy per Country over the years')
ax.set_xlabel('Countries')
ax.set_ylabel('Life expectancy at birth in years')
#ax.set_yticks([40, 50, 60, 70, 80])
ax.set(ylim=(40, 90))

plt.savefig('Life_Expectancy_Over_Years.png')

plt.show()
```

```python
g = sns.FacetGrid(data=df, col='Year', hue='Country', col_wrap=4, height=3, palette='Set1')
g = (g.map(sns.scatterplot, 'GDP', 'LEABY', edgecolor="red").add_legend())
g.set_axis_labels('GDP', 'Life Expectancy')

plt.savefig('Scatter_Plot_over_2000-2015.png')

plt.show()
```

```python
g3 = sns.FacetGrid(df, col="Country", col_wrap=3, height=4)
g3.set(xticks=[2000, 2005, 2010, 2015])
g3 = (g3.map(plt.plot, "Year", "LEABY"))
g3.set_axis_labels('Years', 'Life Expectancy')


plt.savefig('Life_Expectancy_plot.png')

plt.show()
```
